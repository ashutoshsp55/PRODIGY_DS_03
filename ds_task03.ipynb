## Data Visualisation using Matplotlib and Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numerical_features = ['age', 'balance', 'duration', 'campaign', 'pdays', 'previous']

plt.figure(figsize=(8, 4))
sns.histplot(data=df, x='age', bins=20, kde=True)
plt.title(f'Histogram of age')
plt.xlabel('age')
plt.ylabel('Frequency')
plt.show()
```

```python
for feature in numerical_features:
    plt.figure(figsize=(4, 2))
    sns.boxplot(data=df[feature], orient='h')
    plt.title(f'Box of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
categorical_features = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
for feature in categorical_features:
    plt.figure(figsize=(4, 2))
    sns.countplot(data=df, x=feature, order=df[feature].value_counts().index)
    plt.title(f'Bar Plot of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='age',y='balance',alpha=0.5)
plt.title('Scatter Plot of Age vs. Balance')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.show()
```

```python
X = df[selected_columns]
label_encoders = {}
for column in X.select_dtypes(include=['object']).columns:
    label_encoders[column] = LabelEncoder()
    X[column] = label_encoders[column].fit_transform(X[column])
```

## Predicting whether customer will buy the card or not by taking user input

```python

def predict_subscription():
    age = int(input("Enter age: "))
    job = input("Enter job (e.g., management, blue-collar, technician): ")
    marital = input("Enter marital status (e.g., married, single, divorced): ")
    education = input("Enter education level (e.g., primary, secondary, tertiary): ")
    default = input("Enter default status (yes/no): ")
    balance = float(input("Enter account balance: "))
    housing = input("Enter housing loan status (yes/no): ")
    loan = input("Enter personal loan status (yes/no): ")
    contact = input("Enter contact type (e.g., cellular, telephone): ")
    day = int(input("Enter day of the month: "))
    month = input("Enter month (e.g., jan, feb, ...): ")
    duration = int(input("Enter duration of last contact in seconds: "))
    campaign = int(input("Enter number of contacts performed during this campaign: "))
    pdays = int(input("Enter number of days since the customer was last contacted: "))
    previous = int(input("Enter number of contacts performed before this campaign: "))
    poutcome = input("Enter outcome of the previous marketing campaign (e.g., success, failure, nonexistent): ")

    job_encoded = label_encoders['job'].transform([job])[0]
    marital_encoded = label_encoders['marital'].transform([marital])[0]
    education_encoded = label_encoders['education'].transform([education])[0]
    default_encoded = label_encoders['default'].transform([default])[0]
    housing_encoded = label_encoders['housing'].transform([housing])[0]
    loan_encoded = label_encoders['loan'].transform([loan])[0]
    contact_encoded = label_encoders['contact'].transform([contact])[0]
    month_encoded = label_encoders['month'].transform([month])[0]
    poutcome_encoded = label_encoders['poutcome'].transform([poutcome])[0]

    prediction = clf.predict([[age, job_encoded, marital_encoded, education_encoded, default_encoded,
                               balance, housing_encoded, loan_encoded, contact_encoded, day, month_encoded,
                               duration, campaign, pdays, previous, poutcome_encoded]])
    
    
    return prediction[0]


predicted_value = predict_subscription()

print("Predicted value:", predicted_value)
```
